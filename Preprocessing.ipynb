```python
import pandas as pd

df = pd.read_csv("personalized_learning_dataset.csv")
```

```python
df.head()
```

```python
required_cols = [
    "Student_ID",
    "Age",
    "Education_Level",
    "Engagement_Level",
    "Learning_Style"
]
```

```python
df = df[required_cols]
```

```python
df = df.drop_duplicates()

df = df.dropna()
```

```python
df.isnull().sum()
```

```python
summary = df['Learning_Style'].value_counts().reset_index()
summary.columns = ['Learning_Style', 'Number of Posts']
print("Learning Style distribution:\n")
print(summary)
```

```python
df.to_csv("Preprocessed Personalized Dataset.csv", index=False)
```
